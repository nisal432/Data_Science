```python
#Broadcasting rules
# basically if there is 1 on either row or column you can increase it to match with the other array to make both same dimension and perform mathematical operationis
a1 = np.arange(9).reshape(3,3)
a2 = np.arange(3)
print(f'a2 = {a2}')
a1+a2 #a2 shape is only(,3) while a1 shape is (3,3) so for addition both array's dimension must be same now look at magic of broadcasting
#basically if the array is only 1 dimension then first make it 2d by putting 1 in front i.e (1,3) then make it same as the another array that is 3
#if the array is already 2d and doesn't include  1 on either row or column and the shape doesnt match with another array then error will show up

```

```python
#sigmoid basically makes any value in the range of 0 and 1 negative values are close to 0 then positive values are close to 1
# e/exp is just a constant like pi its value is 2.71828......
#basically some understanding of e and pi which is actually cool like how it was formed and why its value is that
print(np.exp(1),np.exp(2)) #basically first argument is it's power exp(2) = exp square
#check
a1 = np.arange(1,10)
1/a1
def sigmoid(arr):
	return 1/1-(np.exp(-arr))
sigmoid(a1)
```

```python
#The slope of a curve at a particular point. is derivative
#Mean Squarred error = actual value - predected value of all the iterations and squarring them so that they don't cancel each other out and adding all and divided all by n which is basically mean
def meanSquarredError(actualArr, predArr):
	return np.mean((actualArr-predArr)**2)
a2 = np.arange(11,20)
print(meanSquarredError(a1,a2))
```
