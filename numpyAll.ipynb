```python
#About slicing
import numpy as np
list1 = [1,2,3,4,5,6,7,8]
list2 = list1
check = list1[::] 
check.pop()
# list2.pop() to check if slicing copies and by common sense it will deep copy
list1
```

```python
a1 = np.arange(10).reshape(2,5)
print(a1)
a1[:, :2]
#slicing 2d arrays 

```

```python
a3 = np.arange(1,28).reshape(3,3,3)
#will try to access/get 1st 2d arr and 3rd 2d arr's first and last column
print(a3)# to check 
a3[::2,:, ::2 ]#done
```

```python
#Fancy Indexing
#on slicing we can access the elements by certain patterns but if we want to access element and there is no pattern then we use fancy indexing
print(a3)
# print(a3[[0,1],[[0],[2]],[0,2]]) trial only failed but discovered something new 
a3[[0,0,0,0,1,1],[0,0,2,2,2,2],[0,2,0,2,0,2]] # i think this is not the way but get's the job done
```

```python
a3<20#basically returns a boolean 
print(a3<20)
a3[(a3<20) & (a3<30)]
#this is basically boolean indexing
```
